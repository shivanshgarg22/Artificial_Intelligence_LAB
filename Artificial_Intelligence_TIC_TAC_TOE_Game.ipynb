```python
board = [' ' for x in range(10)]
def printBoard(board):
    print('   |   |   ')
    print(' ' + board[1] + ' | ' + board[2] + ' | ' + board[3])
    print('   |   |   ')
    print('------------')
    print('   |   |   ')
    print(' ' + board[4] + ' | ' + board[5] + ' | ' + board[6])
    print('   |   |   ')
    print('------------')
    print('   |   |   ')
    print(' ' + board[7] + ' | ' + board[8] + ' | ' + board[9])
    print('   |   |   ')

def IsWinner(b,l):
  return (b[1]== l and b[2]==l and b[3]==l) or (b[4]== l and b[5]==l and b[6]==l) or  (b[7]== l and b[8]==l and b[9]==l) or (b[1]== l and b[4]==l and b[7]==l) or (b[3]== l and b[6]==l and b[9]==l) or (b[2]== l and b[5]==l and b[8]==l) or (b[1]== l and b[5]==l and b[9]==l) or (b[3]== l and b[5]==l and b[7]==l)


def spaceIsFree(pos):
    return board[pos]==' '

def insertLetter(letter,pos):
    board[pos]=letter

def playerMove(symbol):
    run = True
    while run:
        move = input("please select a position to enter the {sym} between 1 to 9 : ".format(sym=symbol))
        try:
            move = int(move)
            if move > 0 and move < 10:
                if spaceIsFree(move):
                    run = False
                    insertLetter(symbol , move)
                else:
                    print('Sorry, this space is occupied')
            else:
                print('please type a number between 1 and 9')

        except:
            print('Please type a number')


```

```python
print("Welcome to Tic-Tac-Toe!")
printBoard(board)

while board.count(' ') > 1:  # Check if there are empty spaces
    print("Player 1 (X):")
    playerMove('X')
    printBoard(board)
    if IsWinner(board, 'X'):
        print("Player 1 (X) wins!")
        break

    if board.count(' ') == 1:
        print("It's a tie!")
        break

    print("Player 2 (O):")
    playerMove('O')
    printBoard(board)
    if IsWinner(board, 'O'):
        print("Player 2 (O) wins!")
        break

if board.count(' ') == 1 and not IsWinner(board, 'X') and not IsWinner(board, 'O'):
    print("It's a tie!")
```

```python
print("Welcome to Tic-Tac-Toe!")
printBoard(board)

while board.count(' ') > 1:  # Check if there are empty spaces
    # Player 1's turn
    print("Player 1 (X):")
    playerMove('X')
    printBoard(board)
    if IsWinner(board, 'X'):
        print("Player 1 (X) wins!")
        break

    if board.count(' ') == 1:
        print("It's a tie!")
        break

    # AI's turn
    print("AI (O):")
    move_made = False

    # AI winning move
    for i in range(1, 10):
        if spaceIsFree(i):
            board[i] = 'O'
            if IsWinner(board, 'O'):
                move_made = True
                break
            board[i] = ' '  # Undo the move

    # Block Player 1's winning move
    if not move_made:
        for i in range(1, 10):
            if spaceIsFree(i):
                board[i] = 'X'
                if IsWinner(board, 'X'):
                    board[i] = 'O'
                    move_made = True
                    break
                board[i] = ' '  # Undo the move

    # Play the center if available
    if not move_made and spaceIsFree(5):
        board[5] = 'O'
        move_made = True

    # Play a corner if available
    if not move_made:
        for i in [1, 3, 7, 9]:
            if spaceIsFree(i):
                board[i] = 'O'
                move_made = True
                break

    # Play any remaining space
    if not move_made:
        for i in range(1, 10):
            if spaceIsFree(i):
                board[i] = 'O'
                move_made = True
                break

    printBoard(board)
    if IsWinner(board, 'O'):
        print("AI (O) wins!")
        break

if board.count(' ') == 1 and not IsWinner(board, 'X') and not IsWinner(board, 'O'):
    print("It's a tie!")
```
